#**Data Dictionary**

- **Age**: Customer Age

- **Job**: Type of job (admin,blue-collar,entrepreneur,housemaid,management,retired,self-employed,services,student,technician,unemployed,unknown)

- **Marital**: Marital status (divorced,married,single,unknown; note: divorced means divorced or widowed)

- **Education**: Education Level(Basic_4_year, Basic_6_year, Basic_9_year, High-School, Illiterate, Professional_course, University_Degree, Unknown)

- **Default**: Has credit in default

- **Balance**: Average yearly balance

- **Housing**: Has housing loan?

- **Loan**: Has personal loan?

- **Contact**: Contact Communication Type (Cellular, Telephone)

- **Day_of_Week**: Last contact day of the week.

- **Month**: Last contact month of year.

- **Duration:** Last contact duration, in seconds (numeric). Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.

- **campaign**: Number of contacts performed during this campaign and for this client (numeric, includes last contact)

- **pdays**: Number of days that passed by after the client was last contacted from a previous campaign (numeric; -1 means client was not previously contacted)

- **previous**: Number of contacts performed before this campaign and for this client.

- **poutcome**: Outcome of the previous marketing campaign (failure,nonexistent,success)

- **Y**: Has the client subscribed a term deposit?

At first, I thought PCA and K-Means would help me segment customers effectively. PCA is great for reducing dimensions, and K-Means is widely used for clustering. So, I started by applying PCA to my data, hoping to simplify it while keeping the most important information. But something felt off. The variance explained by each principal component was increasing too gradually. There was no clear "drop-off" point where I could say, "Okay, these are the most important features." Instead, the variance spread out across too many components. That was my first red flag.

Then, I checked the principal component heatmap. It was messy. No strong patterns, just scattered numbers. That’s when I realized – PCA assumes numerical relationships, but my dataset was full of categorical variables. I had already converted them into 0s and 1s using one-hot encoding, but that doesn’t change the fact that categories don’t have natural numerical relationships. PCA was treating them like continuous numbers, which they weren’t.

Still, I wanted to see if clustering would work. I ran K-Means. The Elbow Method suggested seven clusters, so I used that. The results? Not great. The silhouette score was low. That means my clusters weren’t well-separated. Some were too small, others too spread out. K-Means relies on Euclidean distance, but that’s a poor fit for categorical data. It assumes a straight-line distance makes sense between points, but for categories, it really doesn’t.

At this point, it was clear. The problem wasn’t my clustering method itself—it was the data type. PCA and K-Means struggle with categorical data. One-hot encoding didn’t solve the issue; it just increased the number of dimensions without preserving real relationships. I needed a different approach.

**With professors suggestion:**

So now, I’m switching to K-Prototypes Clustering. This method handles both numerical and categorical data properly. It uses Euclidean distance for numbers and Hamming distance for categories. That means my categorical variables won’t be forced into a numerical structure that doesn’t fit. Instead, the algorithm will recognize similarities the way they actually exist.

This should give me better, more meaningful customer segments. No more arbitrary clusters. No more distorted distances. Just clear, logical groupings that make sense.
