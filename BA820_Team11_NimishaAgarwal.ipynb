#**Data Dictionary**

- **Age**: Customer Age

- **Job**: Type of job (admin,blue-collar,entrepreneur,housemaid,management,retired,self-employed,services,student,technician,unemployed,unknown)

- **Marital**: Marital status (divorced,married,single,unknown; note: divorced means divorced or widowed)

- **Education**: Education Level(Basic_4_year, Basic_6_year, Basic_9_year, High-School, Illiterate, Professional_course, University_Degree, Unknown)

- **Default**: Has credit in default

- **Balance**: Average yearly balance

- **Housing**: Has housing loan?

- **Loan**: Has personal loan?

- **Contact**: Contact Communication Type (Cellular, Telephone)

- **Day_of_Week**: Last contact day of the week.

- **Month**: Last contact month of year.

- **Duration:** Last contact duration, in seconds (numeric). Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.

- **campaign**: Number of contacts performed during this campaign and for this client (numeric, includes last contact)

- **pdays**: Number of days that passed by after the client was last contacted from a previous campaign (numeric; -1 means client was not previously contacted)

- **previous**: Number of contacts performed before this campaign and for this client.

- **poutcome**: Outcome of the previous marketing campaign (failure,nonexistent,success)

- **Y**: Has the client subscribed a term deposit?

## Preliminary Analysis

Our project aims to segment customers and identify anomalies in financial behaviors. While clustering techniques help group customers based on similarity, Association Rule Mining (ARM) provides exploratory insights into relationships between categorical features such as education, job type, and communication methods.

```python
# Categorising Balances
balance_bins = [-np.inf, 0, 1000, 5000, np.inf]
balance_labels = ['Negative', 'Low', 'Medium', 'High']
combined_df['balance_category'] = pd.cut(combined_df['balance'], bins=balance_bins, labels=balance_labels)
```

ChatGPT. (2025). Categorizing customer balances into meaningful groups for analysis [Large language model output]. OpenAI. https://chatgpt.com/share/67b6403d-b380-8002-90b1-317d8b7ba6de

```python
categorical_features = ['balance_category', 'education', 'housing', 'loan', 'contact', 'job', 'month', 'y']
df_transactions = combined_df[categorical_features].copy()

df_transactions = df_transactions.astype(str)
```

```python
from mlxtend.preprocessing import TransactionEncoder

transactions = df_transactions.values.tolist()

encoder = TransactionEncoder()
encoded_array = encoder.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(encoded_array, columns=encoder.columns_)

df_encoded.head()
```

By applying Apriori Algorithm, we sought to uncover patterns in customer demographics, financial behaviors, and marketing responses to help refine our segmentation approach and understand key interactions between features.

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(df_encoded, min_support=0.05, use_colnames=True)

frequent_itemsets
```

The frequent itemsets analysis gives us a clearer picture of the most common financial and demographic patterns in our dataset. The most striking observation is that a majority of customers fall into the low balance category, making up over half of the dataset (51.5%). Medium-balance customers follow at 26%, while 16% of customers have negative balances, meaning a significant portion may be facing financial difficulties. In contrast, only 6.3% of customers have a high balance, indicating that truly high-value clients are rare in this dataset.

Beyond financial standing, we also see that the "admin." job category appears frequently (11.4%), suggesting that administrative professionals make up a notable customer segment. However, no other job categories appear in the top five most frequent patterns, meaning that other professions may be more evenly spread across the dataset rather than concentrated in large numbers.

These findings have important implications for how the bank approaches its customers. Since low-balance customers dominate, financial products and marketing strategies should be designed to appeal to this group, potentially by offering low-risk savings plans, financial literacy resources, or promotions to encourage deposits. Meanwhile, the high proportion of customers with negative balances suggests a need for credit-building programs or debt management solutions to help them regain financial stability.

On the other end of the spectrum, the relatively small number of high-balance customers means that premium banking services should be highly targeted to ensure they reach the right clients. Since administrative professionals form a significant part of the bank’s customer base, the bank could explore whether this group has specific financial needs and create tailored products to serve them better.

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

rules = rules.sort_values("lift", ascending=False)

rules.head(10)
```

The association rule mining results gave us a clearer picture of how different types of customers behave, particularly in terms of their education, job roles, and communication preferences. One of the strongest patterns we found is that customers with tertiary education are highly likely to work in management roles and prefer to be contacted via cellular communication. On the other hand, blue-collar workers tend to have only primary education and are much less associated with higher education levels. This suggests that the bank can take a more personalized approach, offering premium financial products to professionals in management while focusing on financial literacy and more accessible loan options for blue-collar workers.

Another key takeaway is that customers in management, who are often highly educated, are far more reachable through mobile communication channels like calls and SMS. This confirms that digital outreach is the most effective way to engage with high-value customers, making traditional marketing methods like direct mail or branch visits less relevant for this segment.

We also found that blue-collar workers with primary education form a distinct financial demographic that could benefit from tailored financial education and specific loan offerings. Interestingly, high-balance customers did not appear strongly in any rules, suggesting that wealthier customers may have more diverse behaviors that do not fit into clear categorical patterns. This could mean they require a more nuanced, case-by-case approach when designing financial products or marketing strategies.

## Next Steps

To improve customer segmentation and marketing strategies, we can fine-tune Association Rule Mining (ARM). Lowering min_support slightly (e.g., from 0.05 to 0.03) may uncover hidden patterns, while increasing min_confidence helps filter out weaker insights and focus on what truly matters.

We can also analyze balance categories (low, medium, high) separately to see how different customer groups behave. This will help us create more targeted marketing strategies, like offering specific loan products to low-balance customers.

Finally, focusing on lift and leverage values will highlight the strongest and most meaningful patterns. These refinements will help ensure our insights lead to smarter, more effective marketing decisions.
