At first, I thought PCA and K-Means would help me segment customers effectively. PCA is great for reducing dimensions, and K-Means is widely used for clustering. So, I started by applying PCA to my data, hoping to simplify it while keeping the most important information. But something felt off. The variance explained by each principal component was increasing too gradually. There was no clear "drop-off" point where I could say, "Okay, these are the most important features." Instead, the variance spread out across too many components. That was my first red flag.

Then, I checked the principal component heatmap. It was messy. No strong patterns, just scattered numbers. That’s when I realized – PCA assumes numerical relationships, but my dataset was full of categorical variables. I had already converted them into 0s and 1s using one-hot encoding, but that doesn’t change the fact that categories don’t have natural numerical relationships. PCA was treating them like continuous numbers, which they weren’t.

Still, I wanted to see if clustering would work. I ran K-Means. The Elbow Method suggested seven clusters, so I used that. The results? Not great. The silhouette score was low. That means my clusters weren’t well-separated. Some were too small, others too spread out. K-Means relies on Euclidean distance, but that’s a poor fit for categorical data. It assumes a straight-line distance makes sense between points, but for categories, it really doesn’t.

At this point, it was clear. The problem wasn’t my clustering method itself—it was the data type. PCA and K-Means struggle with categorical data. One-hot encoding didn’t solve the issue; it just increased the number of dimensions without preserving real relationships. I needed a different approach.

**With professors suggestion:**

So now, I’m switching to K-Prototypes Clustering. This method handles both numerical and categorical data properly. It uses Euclidean distance for numbers and Hamming distance for categories. That means my categorical variables won’t be forced into a numerical structure that doesn’t fit. Instead, the algorithm will recognize similarities the way they actually exist.

This should give me better, more meaningful customer segments. No more arbitrary clusters. No more distorted distances. Just clear, logical groupings that make sense.

1. Elbow Method

- Train K-Prototypes for a range of k values (e.g., 2 to 15).
- Record the cost (often kproto.cost_) for each k.
- Plot the cost against k.
- Look for an “elbow”—a point where the cost starts decreasing more slowly, indicating diminishing returns for adding more clusters.


```python
import matplotlib.pyplot as plt
from kmodes.kprototypes import KPrototypes

# Convert dataframe to numpy array
data = k_prototype_df.to_numpy()

# Define categorical column indices
categorical_indices = [k_prototype_df.columns.get_loc(c) for c in categorical_cols]

costs = []
K = range(2, 5)  # Adjust range as needed

for k in K:
    kproto = KPrototypes(n_clusters=k, init='Huang', random_state=42, max_iter = 20)
    kproto.fit(data, categorical=categorical_indices)
    costs.append(kproto.cost_)

plt.plot(K, costs, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Cost')
plt.title('Elbow Method for K-Prototypes')
plt.show()
```

# **Background study on Portugal(2008 to 2010)**

**Job Distribution in Portugal (2008–2010)**

Between 2008 and 2010, Portugal's job market was dominated by three main sectors: services, industry, and agriculture. The services sector employed about 60–70% of the workforce, including jobs in retail, banking, healthcare, education, and tourism (Banco de Portugal, 2013). The industrial sector accounted for 20–25% of employment, with strong representation in manufacturing, construction, and energy. Lastly, agriculture, forestry, and fishing provided jobs for approximately 10% of workers, mostly in rural areas (European Central Bank [ECB], 2018).

**Impact of the 2008 Financial Crisis**
The global financial crisis had a significant impact on Portugal. Unemployment rose from 7.6% in 2008 to 10.8% in 2010 (World Bank, 2023). Sectors such as construction, retail, and manufacturing were heavily affected, leading to numerous business closures and job losses. Young professionals faced even greater challenges, with youth unemployment reaching over 20% by 2010 (Statista, 2023). The crisis also led to an increase in temporary contracts and a decrease in stable employment, prompting many workers to shift to informal employment or self-employment (ECB, 2018).

**Salaries and Job Stability**

Salaries varied across industries. Financial services, banking, and public administration offered some of the highest wages. Manufacturing, retail, and tourism sectors provided mid-range salaries, while agriculture and domestic services were at the lower end of the pay scale. The minimum wage was €426 per month in 2008, rising to €475 per month in 2010 (European Commission, 2010). Many jobs paid only slightly above this threshold.

Public sector jobs remained relatively stable compared to the private sector. Government positions in education, healthcare, and administration experienced fewer layoffs. However, small and medium-sized enterprises (SMEs) struggled, leading to increased vulnerability to salary cuts and job losses among private sector workers (Banco de Portugal, 2013).

**Regional Job Distribution**
Job availability varied by region. Lisbon and Porto offered numerous opportunities in finance, technology, and government services. Northern regions, such as Braga and Aveiro, were focused on manufacturing and textiles. Southern and central regions relied heavily on agriculture, tourism, and forestry. The Algarve region, in particular, depended significantly on seasonal tourism jobs, which were often unstable (Banco de Portugal, 2013).

**Banking and Financial Impacts**

The banking industry felt the repercussions of the crisis. Many customers struggled with loan repayments due to job losses, leading to an increase in loan defaults, especially in the construction and retail sectors. Individuals in rural areas relied more on cash transactions and had limited access to banking services, whereas wealthier clients in urban centers were more engaged with financial products (ECB, 2018).

Banks needed to adjust their strategies accordingly. Implementing flexible loan repayment plans became essential. Targeting stable-income professionals, such as government workers, helped mitigate risks. Additionally, there was a heightened demand for savings and budget-friendly products (European Commission, 2010).

- Banks should take advantage of their interest. Offering credit cards, business loans, and home financing will keep them involved. Loyalty programs with cashback rewards or interest discounts will encourage them to stay with the bank. Since they are comfortable exploring new options, introducing investment education programs could push them toward stocks, bonds, or real estate. Keeping them engaged helps build stronger relationships and higher profits.

**Guiding At-Risk Customers Toward Stability**: The third group is young, financially unstable, and struggling with money. Many work in retail, tourism, and construction, sectors that were hit hardest by the crisis. Some have personal loans but low balances, making them more likely to default. They also don’t engage much with banks, which makes it harder to support them.

- Banks need to take a different approach with this group. Instead of pushing credit, they should focus on financial education and debt relief programs. Offering small, secured loans with flexible repayment options can help them build credit safely. Since they don’t respond much to marketing, using AI-driven outreach strategies could help re-engage them. The goal is to help them recover financially while reducing loan risks for the bank.

**Tailoring Services to Different Regions**: Not all regions in Portugal faced the same financial struggles. Lisbon and Porto had more stable jobs in banking and technology, so banks can focus on investment products there. In northern Portugal, where manufacturing is common, banks should provide business loans for small and medium-sized enterprises. The southern and central regions rely on agriculture and tourism, so banks should create seasonal loan plans that match their irregular incomes.By customizing financial products to fit each region’s economy, banks can better serve their customers and improve financial stability across the country.

**Building a Stronger Banking Strategy**: The financial crisis showed that not all customers are the same. Some need investment and savings plans, while others need credit opportunities. Some just need help getting back on their feet. Banks that understand these differences will reduce risk, increase customer engagement, and build stronger relationships.
